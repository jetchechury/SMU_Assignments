# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect, desc
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Precipitation Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
#SELECT prcp
#FROM Measurement
#WHERE Months=12

#precipitation_data
from datetime import datetime

# Calculate the date 1 year ago from the last data point in the database
end_date=session.query(Measurement.date).order_by(Measurement.date.desc()).first()
end_date=end_date[0]
start_date=dt.datetime.strptime(end_date,"%Y-%m-%d")-dt.timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores
precipitation_query=session.query(Measurement.date,Measurement.prcp).\
    filter(Measurement.date>=start_date).all()

#print(precipitation_query)

# Save the query results as a Pandas DataFrame and set the index to the date column
precipitation_df=pd.DataFrame(precipitation_query,columns=['date','precipitaton'])
precipitation_df['date']=pd.to_datetime(precipitation_df['date'])
precipitation_df.set_index('date',inplace=True)
precipitation_df=precipitation_df.dropna(how='any')
precipitation_df.head(25)

#Remove time stamp from start date
start_date=start_date.strftime('%Y-%m-%d')

# Sort the dataframe by date
precipitation_df = precipitation_df.sort_values(by='date',ascending=True)
precipitation_df.head(50)

# Use Pandas Plotting with Matplotlib to plot the data
precipitation_df.plot(figsize=(11.5,9))
plt.title(f"Precipitation in Hawaii from {start_date} to {end_date}")
plt.xlabel("Date")
plt.ylabel("Precipitation (inches)")
plt.savefig("./Images/Precipitation_hawaii.png",bbox_inches="tight")
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data

precipitation_df.describe()
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
station_activity=session.query(Station.name,Measurement.station,func.count(Measurement.id)).\
    filter(Measurement.station==Station.station).group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).all()

station_activity
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?

# Find the most active station:
most_active_station=station_activity[0]
#Unpack tuple
station_name,station_id,row_count=most_active_station[:3]

print(f"The most active stations is {station_name}")



```

```python
#Query temperature data for most active station:
most_active_temp_data=session.query(Measurement.tobs).\
    filter(Measurement.station==station_id).\
    order_by((Measurement.tobs).desc()).all()
```

```python
#Clean data by unpacking temperature tuples
most_active_temp_up = [x for (x), in most_active_temp_data]
most_active_temp_up
```

```python
#Calculate average of temperature of most active station:
from statistics import mean
def Average(most_active_temp_up):
    return mean(most_active_temp_up)
most_active_avgtemp=Average(most_active_temp_up)
```

```python
#Find most active station high temperature and low temperature
most_active_hitemp=most_active_temp_up[0]
most_active_lotemp=most_active_temp_up[-1]

print(f'Information For Most Active Station\n\
-------------------------------------\n\
Name: {station_name}\n\
Highest Temperature: {most_active_hitemp}\n\
Lowest Temperature: {most_active_lotemp}\n\
Average Temperature: {most_active_avgtemp:.2F}')
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
MostActiveStation_TempForYr=session.query(Measurement.tobs).\
    filter(Measurement.date >=start_date,Measurement.station=="USC00519281").\
    order_by((Measurement.tobs).desc()).all()
MostActiveStation_TempForYr
```

```python
MostActiveStation_TempForYr_up= [x for (x), in MostActiveStation_TempForYr]
MostActiveStation_TempForYr_up
```

# Temperature Analysis

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

trip_start_date='2016-11-25'
trip_end_date='2016-12-01'

def calc_temps(trip_start_date,trip_end_date):
   
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= trip_start_date).filter(Measurement.date <= trip_end_date).all()
temp_info=calc_temps(trip_start_date,trip_end_date)
temp_info
```

```python
t_min = temp_info[0][0]
t_avg = temp_info[0][1]
t_max = temp_info[0][2]
t_min,t_avg,t_max
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
plt.bar(1,t_avg,yerr=(t_max-t_min), tick_label='')
plt.title(f'Trip Avgerage Temperature\n{trip_start_date} to {trip_end_date}')
plt.ylabel("Temperature ($^\circ$F)")
plt.xlabel('Hawaii')
plt.savefig("./Images/Trip_Avg_Temp.png",bbox_inches="tight")
plt.show()
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation

total_precp_query=session.query(Measurement.station,Station.name,Station.latitude,Station.longitude,Station.elevation,func.sum(Measurement.prcp)).\
    filter(Measurement.station==Station.station).\
    filter(Measurement.date >= trip_start_date).\
    filter(Measurement.date <= trip_end_date).\
    group_by(Measurement.station).\
    order_by(func.sum(Measurement.prcp).desc()).all()

total_precp_df=pd.DataFrame(total_precp_query,columns=['Station_ID','Station_Name','Latitude','Longitude','Elevation','Total_Precipitation'])
total_precp_df.head(9)
```

## Optional Challenge Assignment

```python
# Test daily_norms: calculate the daily normals for your trip
import datetime as dt

date='11-25'

def daily_norms(date):
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_norms(date)
```

```python
# Set the start and end date of the trip
# Start and end date previously assinged to a variable

# Use the start and end date to create a range of dates
trip_dates=pd.date_range(start=trip_start_date,end=trip_end_date,freq='D')

# Stip off the year and save a list of %m-%d strings
trip_dates_m_d = trip_dates.strftime('%m-%d')

trip_dates_list=list(trip_dates_m_d)
trip_dates_list
```

```python
# Loop through the list of %m-%d strings and calculate the normals for each date
normals=[]
for date in trip_dates_list:
    normals.append(daily_norms(date))
    
normals
```

```python
#Unpack tuples
norm_min = [x[0][0] for x in normals]
norm_avg = [x[0][1] for x in normals]
norm_max = [x[0][2] for x in normals]
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
trip_normals_df=pd.DataFrame()
trip_normals_df['Norm_Min']=norm_min
trip_normals_df['Norm_Avg']=norm_avg
trip_normals_df["Norm_Max"]=norm_max
trip_normals_df["Date"]=trip_dates_list
#Set date as index
trip_normals_df=trip_normals_df.set_index("Date")
trip_normals_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
trip_normals_df.plot.area(stacked=False, figsize=(10,6))

plt.title(f"Trip Daily Normals\n{trip_start_date} to {trip_end_date}")
plt.ylabel("Temperature ($^\circ$F)")

plt.savefig("./Images/trip_normals_hawaii.png",bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import glob
```

```python
# compare headers in monthly sheets
path = r'/Users/jessicaetchechury/Desktop/Data_Science_Bootcamp/GitHub_Repo/Unit_20_Tableau/tripDataUnzipped' # use your path
all_files = glob.glob(path + "/2017*.csv")

last_header = None
# Loop through the files
for file in all_files:
    with open(file) as f:
        # Read the header
        header = f.readline()
        # If this is the first file, then store the
        # header for later comparison
        if last_header == None:
            last_header = header
        # Check that the header match the last header
        elif header != last_header:
            print("\n")
            print(f"{file} does not match ")
            quit()
        # Read the remaining lines
        lines = f.readlines()
```

```python
# read in files that do not have same headers
td_17_01=pd.read_csv('./tripDataUnzipped/201701-citibike-tripdata.csv')
td_17_02=pd.read_csv('./tripDataUnzipped/201702-citibike-tripdata.csv')
td_17_03=pd.read_csv('./tripDataUnzipped/201703-citibike-tripdata.csv')
```

```python
# inspect column headers
print(td_17_01.columns)
print(td_17_02.columns)
print(td_17_03.columns)
```

```python
# concat dataframes
td_17_123=pd.concat([td_17_01, td_17_02,td_17_03], ignore_index=True)
```

```python
# rename columns to match other data sets
td_17_123=td_17_123.rename(columns={'Trip Duration':'tripduration', 'Start Time':'starttime',
                                    'Stop Time':'stoptime', 'Start Station ID':'start station id',
                                    'Start Station Name':'start station name', 
                                    'Start Station Latitude':'start station latitude',
                                    'Start Station Longitude':'start station longitude', 
                                    'End Station ID':'end station id',
                                    'End Station Name':'end station name',
                                    'End Station Latitude': 'end station latitude', 
                                    'End Station Longitude':'end station longitude', 
                                    'Bike ID':'bikeid', 
                                    'User Type':'usertype',
                                    'Birth Year':'birth year', 
                                    'Gender':'gender'})
td_17_123.head()
```

```python
# Concat other files for 2017
import pandas as pd
import glob

path = r'/Users/jessicaetchechury/Desktop/Data_Science_Bootcamp/GitHub_Repo/Unit_20_Tableau/tripDataUnzipped/Group2_2017' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
    print(filename)
    
td_17_allothers = pd.concat(li, axis=0, ignore_index=True)
```

```python
# concat all 2017 data frames
td_17=pd.concat([td_17_123, td_17_allothers], ignore_index=True)

```
